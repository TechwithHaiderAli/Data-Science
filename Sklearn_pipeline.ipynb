```python
import pandas as pd
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

```python
# import titanic dataset from  seaborn

# Import the titanic dataset from seaborn
df = sns.load_dataset('titanic')

# Print the first few rows of the dataset
df.head()
```

```python
#  Select  age sex fare pclass embarked as features and survived as target

X = df[['age', 'sex', 'fare', 'pclass', 'embarked']]
y = df['survived']
```

```python
#  Split into train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#   Divide the features into numerical_features and categorical_features

numerical_features = ['age', 'fare']
categorical_features = ['sex', 'pclass', 'embarked']
```

```python
numeric_transformer=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median'))
    ])
categorical_transformer=Pipeline(steps=[
  ('imputer',SimpleImputer(strategy='most_frequent')),
  ('encoder',OneHotEncoder(handle_unknown='ignore'))
  ])
preprocessor=ColumnTransformer(

    transformers=[
     ('num',numeric_transformer,numerical_features),
     ('cat',categorical_transformer,categorical_features )
 ]
)
```

```python
pipeline=Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier',RandomForestClassifier(random_state=42))
]
)
```

```python
pipeline.fit(X_train,y_train)
```

```python
#  make predictions and evaluate accuracy

y_pred=pipeline.predict(X_test)
accuracy_score(y_test,y_pred)
```
